First of all, I need to read off the input text to create a dataframe. 

```python
import pandas as pd

#CREATING A DATAFRAME FROM THE INPUT FILE
full_input = pd.DataFrame([line.strip().split(' ') for line in open('input.txt', 'r')])
full_input
```

If the condition for the safe reports was only that the levels have to increase or decrease throughout the report, I would probably calculate the running mean in each report and then calculate the differences between them, to see if they are all positive or negative. However, as the change between each level cannot be greater than 3 as well, I will simply parse through the whole dataframe one line at a time. Not a great solution, we'll see how long that takes. 

EDIT FROM THE SECOND PROBLEM:
In this case, we need to add another condition - of a possibility of a bad level, without which the report is otherwise safe. This should be fairly easy to do by splitting the outcomes based on whether the number of bad levels is zero or one.

```python
safe_reports=[1]*len(full_input)

for row in full_input.index:
    counter=0
    trend=0
    bad_levels=0
    column_push=0
    for column in full_input.columns[1:8]:
        if type(full_input.loc[row,column]) == str:
            if bad_levels==0:
                delta=int(full_input.loc[row,column])-int(full_input.loc[row,column-1])

                if abs(delta)>3:
                    bad_levels=1
                    column_push=1
                elif delta==0:
                    bad_levels=1
                    column_push=1
                elif delta>0:
                    trend=trend+1
                else:
                    trend=trend-1

                counter=counter+1

                if counter != abs(trend):
                    bad_levels=1
                    column_push=1
                    counter=counter-1
                    if delta>0:
                        trend=trend-1
                    elif delta<0:
                        trend=trend+1
                
            elif bad_levels==1:
                delta=int(full_input.loc[row,column])-int(full_input.loc[row,column-1-column_push])

                if abs(delta)>3:
                    safe_reports[row]=0
                    break
                elif delta==0:
                    safe_reports[row]=0
                    break
                elif delta>0:
                    trend=trend+1
                else:
                    trend=trend-1

                counter=counter+1

                if counter != abs(trend):
                    safe_reports[row]=0
                    break

                column_push=0
            
        else:
            pass

safe_reports
```

```python
print(sum(safe_reports))
```

However, this method does not take into account the possibility that the first level might be the bad one! Therefore, I should add the first problems code as well and use it on the dataframe without the zeroth column.

```python
new_input=full_input.drop(columns=0)
new_input
```

```python

safe_reports_2=[1]*len(new_input)

for row in new_input.index:
    counter=0
    trend=0
    for column in new_input.columns[1:7]:
        if type(new_input.loc[row,column]) == str:
            delta=int(new_input.loc[row,column])-int(new_input.loc[row,column-1])
            #print(delta)
            if abs(delta)>3:
                safe_reports_2[row]=0
                break
            elif delta==0:
                safe_reports_2[row]=0
                break
            elif delta>0:
                trend=trend+1
            else:
                trend=trend-1

            counter=counter+1

            if counter != abs(trend):
                safe_reports_2[row]=0
                break
        else:
            pass

safe_reports_2
```

```python
print(sum(safe_reports_2))
```

```python
full_input['First set of safe']=safe_reports
full_input['Second set of safe']=safe_reports_2
full_input
```

```python
total_safe=full_input.loc[:,['First set of safe','Second set of safe']].any(axis=1)
print(sum(total_safe))
```
